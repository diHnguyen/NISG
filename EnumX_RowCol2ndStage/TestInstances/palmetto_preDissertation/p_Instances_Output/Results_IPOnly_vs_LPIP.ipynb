```python
new_df = pd.merge(df_IP, df_LP,  how='left', left_on=['dataSet','ins'], right_on = ['dataSet','ins'])
new_df = pd.merge(new_df, df_SVI,  how='left', left_on=['dataSet','ins'], right_on = ['dataSet','ins'])
new_df['timerow_IP_LP'] = (new_df['t_RowGen_IP'] - new_df['t_RowGen_LP'])/new_df['t_RowGen_IP']
new_df['timediff_IP_LP'] = (new_df['totalTime_IP'] - new_df['totalTime_LP'])/new_df['totalTime_IP']
new_df['timerow_IP_SVI'] = (new_df['t_RowGen_IP'] - new_df['t_RowGen_SVI'])/new_df['t_RowGen_IP']
new_df['timediff_IP_SVI'] = (new_df['totalTime_IP'] - new_df['totalTime_SVI'])/new_df['totalTime_IP']

new_df= new_df.sort_values(by=['dataSet', 'ins'])
new_df.to_csv('./allOutputs.csv') 

#Checking N30d10
# new_df[new_df.dataSet == 'N30_d10_']['timediff_IP_SVI'].sum()
new_df[['dataSet','n_RowGen_IP', 't_RowGen_IP','n_RowGen_LP','t_RowGen_LP','n_RowGen_SVI','t_RowGen_SVI']]
```

```python
new_df['weight_IP_LP'] = np.where(new_df.timerow_IP_LP.notnull(), 1, 0)
display(new_df.style.format({'timediff_IP_LP': '{:,.1%}'.format, 'timediff_IP_SVI': '{:,.1%}'.format,
                            'timerow_IP_LP': '{:,.1%}'.format, 'timerow_IP_SVI': '{:,.1%}'.format}))
```

```python
df_Summary1 = new_df.groupby(['dataSet', 'weight_IP_LP'], as_index=False).aggregate({'totalTime_IP':np.average, 'timerow_IP_LP':np.average,'timediff_IP_LP':np.average})#,'timecol_IP_SVI':np.average, 'timediff_IP_SVI':np.average })
df_Summary1 = df_Summary1.dropna()
df_Summary1 = df_Summary1.drop(['weight_IP_LP'], axis=1)
df_Summary1
```

```python
df_Summary2 = new_df.groupby(['dataSet'], as_index=False).aggregate({'totalTime_IP':np.average, 'timerow_IP_SVI':np.average,'timediff_IP_SVI':np.average})#,'timecol_IP_SVI':np.average, 'timediff_IP_SVI':np.average })
df_Summary2 = df_Summary2.drop(['totalTime_IP'], axis=1)
df_Summary2.dropna()
df_Summary2
```
