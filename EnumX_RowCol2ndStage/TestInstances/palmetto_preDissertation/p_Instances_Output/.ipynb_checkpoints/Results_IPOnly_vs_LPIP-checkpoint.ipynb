```python
import re
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import ast
```

```python
location = "./IP_Only/"

#"./Documents/GitHub/UncertainTarget/EnumX_RowCol2ndStage/TestInstances/"
directory = os.fsencode(location)

# def importSet(filename, setRow, setLabel):
df_IP = pd.DataFrame(columns=('dataSet', 'ins','theta', 'totalTime', 't_RowGen', 'n_RowGen', 't_ColGen','n_ColGen'))
# location = "Documents/GitHub/UncertainTarget/EnumX_RowCol2ndStage/TestInstances/p_Instances_Output/"
allSets = []
k = 0
for file in os.listdir(directory):
    filename = os.fsdecode(file) 
#     print(filename)
    if filename.startswith('O_'):
        filePath = os.path.join(location, filename)
#         print(filePath)
        thisFile = open(filePath, "r")
          # remove /n at the end of each line
        lines_E = thisFile.readlines()
        thisFile.close()
        r = 0
        df_row = [0 for i in range(8)]
        for index, line in enumerate(lines_E):
            lines_E[index] = line.strip()
        for line in lines_E:
#             print(line)
            line = line.replace(', ', ',')
            line = line.replace('; ', ' ')
            elem = line.split(' = ')
            if elem[0] == "dataSet":
#                 
                df_row[0] = elem[1]
            elif elem[0] == "ins":
                df_row[1] = elem[1]
            elif elem[0] == "theta":
                df_row[2] = elem[1]
            elif elem[0] == "totalTime":
                df_row[3] = elem[1]
            elif elem[0] == "t_RowGen":
                df_row[4] = elem[1]
            elif elem[0] == "n_RowGen":
                df_row[5] = elem[1]
            elif elem[0] == "t_ColGen":
                df_row[6] = elem[1]
            elif elem[0] == "n_ColGen":
                df_row[7] = elem[1]
                        
#             r = r+1
#         print(df_row)
        k=k+1
        df_IP.loc[k] = df_row
for colName in ['ins','theta', 'totalTime', 't_RowGen', 'n_RowGen','t_ColGen','n_ColGen']:
    df_IP[colName] = pd.to_numeric(df_IP[colName])
#         print(df_E)
# display(df_E.style.format({'RootOptGap': '{:,.1%}'.format}))
# display()
df_IP.groupby(['dataSet'], as_index=False).aggregate({'totalTime':np.average,'t_RowGen':np.average,'t_ColGen':np.average, 't_ColGen':np.average })
```
