```julia
using JuMP
using Gurobi
```

```julia
global C = -2 #100
global v = 10 #50
# global arcs = [1 2; 1 3; 2 3]
global arcs = [1 2; 1 3; 1 4; 2 3; 2 5; 3 5; 4 5]
# global paths = [[1 2 3],
#             [1 3]] 
global paths = [[1 2 3 5],
                [1 2 5],
                [1 3 5],
                [1 4 5]] 
# D = zeros(7,4)
# D[1,1] = 1
# D[4,1] = 1
# D[6,1] = 1
# D[1,2] = 1
# D[5,2] = 1
# D[2,3] = 1
# D[6,3] = 1
# D[3,4] = 1
# D[7,4] = 1

global N = collect(1:maximum(arcs))
origin = 1
destination = 5 #3
global arcNum = length(arcs[:,1])

function ConstructQ()
    global arcs # = [1 2; 1 3; 1 4; 2 3; 2 5; 3 5; 4 5]
    global v
    global C
    global paths

    Q = ones((length(arcs[:,1]), length(paths))).*v
    D = zeros((length(arcs[:,1]), length(paths)))
    
    for l = 1:length(paths)
        for k = 1:length(paths[l])-1
            arc_on_path_i = paths[l][k]
            arc_on_path_j = paths[l][k+1]
            arcSet = findall(arcs[:,1].==arc_on_path_i)
            for i in arcSet   
                if arcs[i,2] == arc_on_path_j
                   Q[i,l] = C 
                    D[i,l] = 1
                end
            end
        end
    end
    return Q,D
end

function MaxFlow(x, origin, destination)
    global arcs
    global N
    global arcNum
    global numPaths
    global Q, D

    
#     for i = 1:arcNum
#        Q[i,1] = -2 
#     end
    
    println("Q = ", Q)
    println("D = ", D)
    
#     println("arcs = ", arcs)
#     println("N = ", N)
    gurobi_env = Gurobi.Env()
    setparams!(gurobi_env, Heuristics=0.0, Cuts = 0, OutputFlag = 0)
    m = Model(() -> Gurobi.Optimizer(gurobi_env))

    @variable(m, y[1:arcNum]>=0)
    @variable(m, u)
#     flow_cons = 
    flow_cons = Array{JuMP.ConstraintRef}(undef, length(N)-1)
    ucap_cons = Array{JuMP.ConstraintRef}(undef, arcNum)
    xcap_cons = Array{JuMP.ConstraintRef}(undef, arcNum)
    conNum = 1
    
    
#     b_flow = zeros(total_constr_num)
#     b_flow[1] = 1
#     b_ucap = 
    #Flow constraints:
    flow_cons[conNum] = @constraint(m, sum(y[i] for i in findall(arcs[:,1].==origin)) == 1)
    conNum = conNum + 1
    for node in N
        if node != origin && node !=destination
            incoming = findall(arcs[:,2].==node)
            outgoing = findall(arcs[:,1].==node)
            flow_cons[conNum] = @constraint(m, sum(y[i] for i in outgoing)-sum(y[i] for i in incoming) == 0)
            conNum = conNum + 1
        end
    end
#     println("2")
    for i = 1:arcNum
        xcap_cons[i] = @constraint(m,  (1-x[i]) >= y[i])
#         b_vector[conNum] = x[i]
#         conNum = conNum + 1
    end
    for i = 1:arcNum
        ucap_cons[i] = @constraint(m,  u >= y[i])
# # #         b_vector[conNum] = x[i]
# # #         conNum = conNum + 1
    end
#     for i = 1:arcNum
#         constr[conNum] = @constraint(m,  y[i] >=0)
#         conNum = conNum + 1
#     end
    
#     println(1)
    @objective(m, Min, u)
#     @objective(m, Min, sum(y[i] for i = 1:arcNum))
#     @objective(m, Max, y[1]+y[2])
    optimize!(m)
    println(m)
    current_Obj = JuMP.objective_value.(m)
    # current_x = JuMP.value.(x)
    current_y = JuMP.value.(y)
#     current_u = JuMP.value.(u)
    flow_dual = zeros(length(N)-1)
    ucap_dual = zeros(arcNum)
    xcap_dual = zeros(arcNum)
#     println("4")
    for i = 1:length(N)-1
        flow_dual[i] = JuMP.dual(flow_cons[i])
    end
    for i = 1:arcNum
        xcap_dual[i] = JuMP.dual(xcap_cons[i])
#         ucap_dual[i] = JuMP.dual(ucap_cons[i])
    end
    println("y_sol = ", current_y)
    println("SP Obj = ",current_Obj)
    println("flow_dual = ", flow_dual)
    println("xcap_dual = ", xcap_dual)
    println("ucap_dual = ", ucap_dual)
#     println("Dual sol = ", current_z)
#     println("Dual mult (b) = ", b_vector)
    return current_Obj, flow_dual, xcap_dual,0 #current_z #, b_vector, total_constr_num
end

global M = Int64(10)
Q,D = ConstructQ()
global numPaths = length(Q[1,:])
println("\nQ")
for i = 1:arcNum
    println(Q[i,:])
end
println("\nD")
for i = 1:arcNum
    println(D[i,:])
end
x_now = zeros(arcNum)
x_now[1] = 1
current_Obj, dualSol, xcap_dual, multiplier_x  = MaxFlow(x_now, origin, destination)
# LPMax(zeros(arcNum), origin, destination)
    
# function MP()
    gurobi_env = Gurobi.Env()
    setparams!(gurobi_env, Heuristics=0.0, Cuts = 0, OutputFlag = 0)
    MP = Model(() -> Gurobi.Optimizer(gurobi_env))

    @variable(MP, x[1:arcNum], Bin)
    @variable(MP, w >= C)
    # @variable(m, y[1:arc_num]>=0)
    @objective(MP, Min, w)
    
    path_cons = Array{JuMP.ConstraintRef}(undef, numPaths)
    @constraint(MP, sum(x[i] for i=1:arcNum) == 1)

    x_now = zeros(arcNum)
    # println("2")
    last_MP = v
    for iter = 1:3
        println("\n\niter = ", iter)
#         current_Obj, dual1, dualSol, xcap_dual, multiplier_x  = MaxFlow(x_now, origin, destination)
#     println("0")
    #     println("b_vector = ", b_vector)
    #     println("total_constr_num : ", total_constr_num)
        arc_ind = length(N)-1
        println("1")
        
#         println("Flow Duals ", flow_dual)
#         println("xCap Duals ", xcap_dual)
    #     println("SP Obj = ",current_Obj)
    #     println("Dual sol = ", current_z)
    #     println("Dual mult (b) = ", b_vector)
#         minCut = findall(xcap_dual.>0)

    #     minCut = minCut[minCut .> arc_ind]
#         println("minCut = ", minCut)
#         @constraint(MP, w <= flow_dual[1]#*b_vector[1] 
#                         + sum(C*xcap_dual[i]*(x[i]-1) for i in minCut)
#                         + sum(v*(1-xcap_dual[i])*(x[i]-1) for i in minCut) )
#         @constraint(MP, w >= dual1#*b_vector[1] 
#                         + sum(xcap_dual[i]*(x[i]-1) for i = 1:arcNum) )
#         @constraint(MP, w >= (dual1
#                         + sum(xcap_dual[i]*(1-x[multiplier_x[i]]) for i = 1:length(xcap_dual))) )
        
        optimize!(MP)
        println(MP)
        
        MP_Obj = JuMP.objective_value.(MP)
    #     current_w = JuMP.value.(w)
        x_now = JuMP.value.(x)
        println("Obj = ",MP_Obj)
        println("x = ", x_now)
        
#         println("subProblem Obj = ", y_cut'*Q[:,k])
    #     println("5")
    y_cut = zeros(arcNum) #[0.5,0.5,0]
    if iter == 1
        for i in [1,2,3]
            y_cut[i] = 1/3
        end
        
        for l = 1:numPaths
            path_cons[l] = @constraint(MP, w >= sum(y_cut[i]*(-M*sum(x[k]*D[k,l] for k = 1:arcNum)  + Q[i,l]  )  for i = 1:arcNum)  )
        end
    end
    if iter == 2
       for i in [2,3]
            y_cut[i] = 1/2
        end
        
        for l = 1:numPaths
#             @constraint(MP, w >= sum(y_cut[i]*(sum(-M*x[k]*D[k,l] for k = 1:arcNum)  + Q[i,l]  )  for i = 1:arcNum)  )
            k = 1
            while x_now'*D[:,k] > 0
                k = k + 1
            end
            JuMP.setRHS(path_cons[l], y_cut'*Q[:,k])
        end 
    end
        println("y*Q = ", y_cut'*Q)
        k = 1
        while x_now'*D[:,k] > 0
            k = k + 1
        end
        println("subProblem Obj = ", y_cut'*Q[:,k])
        if last_MP - MP_Obj <= 1e-6
           break 
        end
        last_MP = MP_Obj

    end

# end

```
