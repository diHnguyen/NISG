```julia
global C = -2 #100
global v = 10 #50

# global arcs = [1 2; 1 3; 2 3]
# global paths = [[1 2 3],
#             [1 3]] 
global arcs = [1 2; 1 3; 1 4; 2 3; 2 5; 3 5; 4 5]
global paths = [[1 2 3 5],
                [1 2 5],
                [1 3 5],
                [1 4 5]] 

global N = collect(1:maximum(arcs))
origin = 1
destination = 3#5
global arcNum = length(arcs[:,1])

function ConstructQ()
    global arcs # = [1 2; 1 3; 1 4; 2 3; 2 5; 3 5; 4 5]
    global v
    global C
    global paths

    Q = ones((length(arcs[:,1]), length(paths))).*v
    D = zeros((length(arcs[:,1]), length(paths)))
    
    for l = 1:length(paths)
        for k = 1:length(paths[l])-1
            arc_on_path_i = paths[l][k]
            arc_on_path_j = paths[l][k+1]
            arcSet = findall(arcs[:,1].==arc_on_path_i)
            for i in arcSet   
                if arcs[i,2] == arc_on_path_j
                   Q[i,l] = -C 
                    D[i,l] = 1
                end
            end
        end
    end
    return Q,D
end

function LPMax(x, origin, destination)
    global arcs
    global N
    global arcNum
    global Q

    Q,D = ConstructQ()
#     for i = 1:arcNum
#        Q[i,1] = -2 
#     end
    numPaths = length(Q[1,:])
    println("Q = ", Q)
    println("D = ", D)
    gurobi_env = Gurobi.Env()
    setparams!(gurobi_env, Heuristics=0.0, Cuts = 0, OutputFlag = 0)
    m = Model(() -> Gurobi.Optimizer(gurobi_env))

    @variable(m, z[1:numPaths] >=0)
#     @variable(m, slack >= 0)
    @variable(m, r)
#     flow_cons = Array{JuMP.ConstraintRef}(undef, length(N)-1)
#     ucap_cons = Array{JuMP.ConstraintRef}(undef, arcNum)
    xcap_cons = Array{JuMP.ConstraintRef}(undef, arcNum*numPaths)
    
    duals = Array{JuMP.ConstraintRef}(undef, arcNum)
#     duals_pos = Array{JuMP.ConstraintRef}(undef, arcNum)
    dual1 = @constraint(m, 1 == sum(z[i] for i = 1:numPaths) )
#     @constraint(m, slack == 0)
    for i = 1:arcNum
       duals[i] = @constraint(m,  sum(Q[i,j]*z[j] for j = 1:numPaths) >= r) 
    end
#     for i = 1:numPaths
#         duals_pos[i] = @constraint(m, 0 <= z[i])
#     end
    conNum = 0
    multiplier_x = []
    for i = 1:arcNum
#         if x[i] != 0
            for j = 1:numPaths
                if D[i,j] == 1
                    conNum = conNum+1
                    xcap_cons[conNum] = @constraint(m,  (1-x[i]) >= z[j])   
                    push!(multiplier_x, i)
                end
            end
#         end
    end
    println("multiplier_x", multiplier_x)
    @objective(m, Max, r)
    optimize!(m)
    println(m)
    current_Obj = JuMP.objective_value.(m)
#     println(JuMP.RHS(m))
    current_z = JuMP.value.(z)
    current_r = JuMP.value.(r)
    dualSol = zeros(arcNum)
#     dualPos = zeros(numPaths)
#     flow_dual = zeros(length(N)-1)
    xcap_dual = zeros(conNum)

#     for i = 1:length(N)-1
#         flow_dual[i] = JuMP.dual(flow_cons[i])
#     end
    println("conNum = ", conNum)
    for i = 1:conNum#numPaths
        xcap_dual[i] = JuMP.dual(xcap_cons[i])
    end
    for i = 1:arcNum
        dualSol[i] = JuMP.dual(duals[i])
    end
#     for i = 1:numPaths
#         dualPos[i] = JuMP.dual(duals_pos[i])
#     end
    println("z_sol = ", current_z)
    println("SP Obj = ",current_Obj)
    println("dual1 = ", JuMP.dual(dual1))
    println("dualSol = ", dualSol)
#     println("dualPos = ", dualPos)
    println("xcap_dual = ", xcap_dual)
   
#     println("yQ = ", current_y'*Q)
    println("yQz = ", dualSol'*Q*current_z)
    return current_Obj, JuMP.dual(dual1), dualSol, xcap_dual, multiplier_x
end

x_now = zeros(arcNum)
# x_now[1] = 1
current_Obj, dual1, dualSol, xcap_dual, multiplier_x  = LPMax(x_now, origin, destination)
```

```julia
function MaxFlow(x, origin, destination)
    global arcs
    global N
    global arcNum
    u = 10
#     println("arcs = ", arcs)
#     println("N = ", N)
    gurobi_env = Gurobi.Env()
    setparams!(gurobi_env, Heuristics=0.0, Cuts = 0, OutputFlag = 0)
    m = Model(() -> Gurobi.Optimizer(gurobi_env))

    @variable(m, y[1:arcNum]>=0)
#     @variable(m, u)
#     flow_cons = 
    flow_cons = Array{JuMP.ConstraintRef}(undef, length(N)-1)
    ucap_cons = Array{JuMP.ConstraintRef}(undef, arcNum)
    xcap_cons = Array{JuMP.ConstraintRef}(undef, arcNum)
    conNum = 1
    
#     b_flow = zeros(total_constr_num)
#     b_flow[1] = 1
#     b_ucap = 
    #Flow constraints:
    flow_cons[conNum] = @constraint(m, sum(y[i] for i in findall(arcs[:,1].==origin)) == u) #1)
    conNum = conNum + 1
    for node in N
        if node != origin && node !=destination
            incoming = findall(arcs[:,2].==node)
            outgoing = findall(arcs[:,1].==node)
            flow_cons[conNum] = @constraint(m, sum(y[i] for i in outgoing)-sum(y[i] for i in incoming) == 0)
            conNum = conNum + 1
        end
    end
#     println("2")
    for i = 1:arcNum
        xcap_cons[i] = @constraint(m,  (1-x[i])*u >= y[i])
#         b_vector[conNum] = x[i]
#         conNum = conNum + 1
    end
#     for i = 1:arcNum
#         ucap_cons[i] = @constraint(m,  u >= y[i])
# #         b_vector[conNum] = x[i]
# #         conNum = conNum + 1
#     end
#     for i = 1:arcNum
#         constr[conNum] = @constraint(m,  y[i] >=0)
#         conNum = conNum + 1
#     end
    
#     println(1)
#     @objective(m, Max, u)
    @objective(m, Max, y[1]+y[2])
    optimize!(m)
    println(m)
    current_Obj = JuMP.objective_value.(m)
    # current_x = JuMP.value.(x)
    current_y = JuMP.value.(y)
    current_u = JuMP.value.(u)
    flow_dual = zeros(length(N)-1)
#     ucap_dual = zeros(arcNum)
    xcap_dual = zeros(arcNum)
#     println("4")
    for i = 1:length(N)-1
        flow_dual[i] = JuMP.dual(flow_cons[i])
    end
    for i = 1:arcNum
        xcap_dual[i] = JuMP.dual(xcap_cons[i])
    end
    println("y_sol = ", current_y)
    println("SP Obj = ",current_Obj)
#     println("Dual sol = ", current_z)
#     println("Dual mult (b) = ", b_vector)
    return current_Obj, flow_dual, xcap_dual #current_z #, b_vector, total_constr_num
end
x_now = zeros(arcNum)
x_now[1] = 1
MaxFlow(x_now, origin, destination)
```

```julia
function LPMin(x, origin, destination)
    global arcs
    global N
    global arcNum
    global Q
    println("1")
    Q, D = ConstructQ()
#     for i = 1:arcNum
#        Q[i,1] = -2 
#     end
    numPaths = length(Q[1,:])
    println("Q = ", Q)
    gurobi_env = Gurobi.Env()
    setparams!(gurobi_env, Heuristics=0.0, Cuts = 0, OutputFlag = 0)
    m = Model(() -> Gurobi.Optimizer(gurobi_env))

    @variable(m, y[1:arcNum]>=0)
    @variable(m, u)
    @variable(m, uC)
    @variable(m, uV)
#     flow_cons = Array{JuMP.ConstraintRef}(undef, length(N)-1)
#     ucap_cons = Array{JuMP.ConstraintRef}(undef, arcNum)
    xcap_cons = Array{JuMP.ConstraintRef}(undef, arcNum)
    conNum = 1
    duals = Array{JuMP.ConstraintRef}(undef, numPaths)
    dual1 = @constraint(m, sum(y[i] for i =1:arcNum) ==1)
#     for j = 1:numPaths
#        duals[j] = @constraint(m, u >= sum(Q[i,j]*y[i] for i = 1:arcNum)) 
#     end
    
    for j = 1:numPaths
        tempQ = D.*C
        @constraint(m, uC >= sum(tempQ[i,j]*y[i] for i = 1:arcNum)) 
    end
    for j = 1:numPaths
        tempQ = (ones(arcNum, numPaths)-D).*v
        @constraint(m, uV >= sum(tempQ[i,j]*y[i] for i = 1:arcNum)) 
    end
    @constraint(m, u >= uV+uC)
    for i = 1:arcNum
#         xcap_cons[i] = @constraint(m, u-x[i]>= y[i])  
        xcap_cons[i] = @constraint(m, 1-x[i]>= y[i])  
    end
#     @constraint(m, 1>=u)
    @objective(m, Min, u)#*C + (sum(y[i] for i = 1:arcNum))*v)
    optimize!(m)
    println(m)
    
    current_Obj = JuMP.objective_value.(m)
    current_y = JuMP.value.(y)
    current_u = JuMP.value.(u)
    dualSol = zeros(numPaths)
#     flow_dual = zeros(length(N)-1)
    xcap_dual = zeros(arcNum)
    
#     for i = 1:length(N)-1
#         flow_dual[i] = JuMP.dual(flow_cons[i])
#     end
    for i = 1:arcNum
        xcap_dual[i] = JuMP.dual(xcap_cons[i])
    end
    for i = 1:numPaths
#         dualSol[i] = JuMP.dual(duals[i])
    end
    println("uC = ", JuMP.value.(uC))
    println("uV = ", JuMP.value.(uV))
    println("u = ", current_u)
    println("y_sol = ", current_y)
    println("SP Obj = ",current_Obj)
    println("dual1 = ", JuMP.dual(dual1))
    println("dualSol = ", dualSol)
    println("xcap_dual", xcap_dual)
    println("yQ = ", current_y'*Q)
    println("yQz = ", current_y'*Q*dualSol)
    return current_Obj, JuMP.dual(dual1), dualSol, xcap_dual
end

x_now = [0,0,0]
obj, dual1, dualSol, xcap_dual = LPMin(x_now, origin, destination)
```
